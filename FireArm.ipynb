```python
import pandas as pd

df = pd.read_csv('fire_arms.csv')

df.head()

neighborhoods = df[['neighborhood', 'total_count']]

dictionary = {}

for row in neighborhoods.iterrows():
    #print(row[1]['neighborhood'], row[1]['total_count'])
    if row[1]['neighborhood'] not in dictionary.keys():
        dictionary[row[1]['neighborhood']] = row[1]['total_count']
    else:
        dictionary[row[1]['neighborhood']] = dictionary.get(row[1]['neighborhood']) + row[1]['total_count']

dictionary_sorted = sorted(dictionary.items(), key=lambda x:x[1], reverse=True)
print(dictionary_sorted)

new_dict = {}

for i in range(1, len(dictionary_sorted)):
    new_dict[dictionary_sorted[i][0]] = dictionary_sorted[i][1]

keys = new_dict.keys()
values = new_dict.values()

new_df = pd.DataFrame({'Neighborhood': keys, 'Number of Firearm Seizures':values})

#ax = new_df.plot.bar(x='neighborhood', y='total_count', rot=0)

chart = new_df.plot.bar(x = 'Neighborhood', y = 'Number of Firearm Seizures', title = 'Pittsburgh Neighborhood Crime', figsize = (30, 10))
```
